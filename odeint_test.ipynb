```python
import numpy as np
import torch
from torchdiffeq import odeint
import matplotlib.pyplot as plt
```

```python
t = torch.linspace(0, torch.pi/2)
f = lambda t, x: torch.cos(t)

y = odeint(f, torch.tensor([0], dtype=torch.float64), t)

plt.plot(t, y)
plt.plot(t, f(t, t))
plt.plot(t[:-1], torch.diff(y, dim=0)[:, 0] / torch.abs(t[1] - t[0]))
plt.legend(['y', 'dy/dt', 'diff'])
plt.show()
```

```python
torch.max(y)
```
